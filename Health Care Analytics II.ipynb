```python
df2 = pd.merge(df,City,on='case_id', how='inner')
df2.head()
```

```python
X = df2.drop('Stay', axis=1)
y= df2.Stay
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X.select_dtypes(include='object').columns
```

```python
X_label = X.select_dtypes(include='object').apply(LabelEncoder().fit_transform)
X.drop(['case_id','Hospital_type_code', 'Hospital_region_code', 'Department', 'Ward_Type',
       'Ward_Facility_Code', 'Type of Admission', 'Severity of Illness',
       'Age'],axis=1, inplace=True)
X.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
bestfeatures = SelectKBest(score_func=chi2,k=8)
fit= bestfeatures.fit(X,y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featurescores= pd.concat([dfcolumns,dfscores], axis=1)
featurescores
```

```python
X = pd.concat([X, X_label], axis=1)
X.head()
```

```python
Admission = pd.DataFrame(X.City_Code_Patient)
X.drop('City_Code_Patient', axis=1, inplace=True)
X = pd.concat([X, Admission], axis=1)
X.drop('patientid', axis=1, inplace=True)
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
X = pd.DataFrame(X_scaled, columns = X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split
```
